```python
#Funciones
def ReplaceList(listOri,strRep,strOri="_Suscriptores"):
    b=[]
    for i in listOri:
        b.append(i.replace(strOri,strRep ))
    return b
def GraficoBarra(Tabla,Clases,Indice,titulo,aggF=np.sum,div=12):
    table = pd.pivot_table(Tabla, values=Clases, index=Indice,aggfunc=aggF)/div
    grafica=px.bar(table, x=table.index, y=Clases, title=titulo)
    return grafica
def ValorPorPersona(Tabla,ColumnaPatron,PatronColumnaA, PatronColumnaB, PatronResultado):
    for i in ColumnaPatron:
        ii=i.replace(PatronColumnaA,PatronColumnaB)
        iii=i.replace(PatronColumnaA,PatronResultado)
        Tabla[iii]=Tabla[ii]/Tabla[i]
    return Tabla
def PrecioKilovatio(Tabla,ColumnaPatron,PatronColumnaA, PatronColumnaB, PatronResultado):
    for i in ColumnaPatron:
        ii=i.replace(PatronColumnaA,PatronColumnaB)
        iii=i.replace(PatronColumnaA,PatronResultado)
        Tabla[iii]=Tabla[ii]/Tabla[i]
    return Tabla


```

```python
# Enriqueser datos
ClasesSuscriptores=["Estrato 1_Suscriptores","Estrato 2_Suscriptores", "Estrato 3_Suscriptores", "Estrato 4_Suscriptores", "Estrato 5_Suscriptores", "Estrato 6_Suscriptores", "Industrial_Suscriptores", "Comercial_Suscriptores", "Oficial_Suscriptores", "Otros_Suscriptores"]
ClasesConsumo=ReplaceList(ClasesSuscriptores,"_Consumo")
ClasesValor=ReplaceList(ClasesSuscriptores,"_Valor Consumo")
ClasesConsumoPorSus=ReplaceList(ClasesSuscriptores,"_Consumo Por Suscritor")
ClasesValorPorSus=ReplaceList(ClasesSuscriptores,"_Valor Por Suscritor")
ClasesPrecioPorWat=ReplaceList(ClasesSuscriptores,"_Precio por Vatio")


F=ValorPorPersona(FinalM,ClasesSuscriptores,"_Suscriptores","_Consumo","_Consumo Por Suscritor")
F=ValorPorPersona(F,ClasesSuscriptores,"_Suscriptores","_Valor Consumo","_Valor Por Suscritor")
F=PrecioKilovatio(F,ClasesConsumoPorSus,"_Consumo Por Suscritor","_Valor Por Suscritor","_Precio por Vatio")
```
